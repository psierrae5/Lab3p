```python
# -------------------------------
# Lab 3 Spint Python: Entrenamiento 2
# Análisis de eventos deportivos
# -------------------------------

# Paso 1: Importar librerías
import pandas as pd

# Paso 2: Cargar los datos desde CSV con rutas absolutas
#ruta = "C:/Users/Usuario01/Downloads/Sprint Python/Lab3/Entrenamiento 2/analisis_datos/"
ruta = "/workspace/"
eventos = pd.read_csv(ruta + "eventos.csv")
aficionados = pd.read_csv(ruta + "aficionados.csv")
promociones = pd.read_csv(ruta + "promociones.csv")

# Verificar primeros registros (opcional)
print("Eventos:\n", eventos.head())
print("Aficionados:\n", aficionados.head())
print("Promociones:\n", promociones.head())

# Paso 3:  Limpieza de datos: eliminar filas con valores nulos
eventos = eventos.dropna()
aficionados = aficionados.dropna()
promociones = promociones.dropna()

# Paso 4: Combinación de DataFrames usando id_evento
df_eventos_aficionados = pd.merge(eventos, aficionados, on="id_evento", how="inner")
df_completo = pd.merge(df_eventos_aficionados, promociones, on="id_evento", how="inner")

# Paso 5: Análisis de datos

# a) Evento con mayor asistencia total
evento_mayor_asistencia = df_completo.loc[df_completo["asistentes_totales"].idxmax()]
print("\nEvento con mayor asistencia total:")
print(evento_mayor_asistencia[["nombre_evento", "asistentes_totales"]])

# b) Rango de edad que asiste más frecuentemente
bins = [0, 18, 30, 45, 60, 100]
labels = ["0-18", "19-30", "31-45", "46-60", "61+"]
df_completo["rango_edad"] = pd.cut(df_completo["edad"], bins=bins, labels=labels)

rango_mas_frecuente = df_completo["rango_edad"].value_counts().idxmax()
print("\nRango de edad más frecuente en los eventos:", rango_mas_frecuente)

# c) Medio publicitario con mayor impacto
df_completo["impacto_promocion"] = df_completo["asistentes_totales"] / df_completo["presupuesto"]

impacto_por_medio = df_completo.groupby("medio_publicitario")["impacto_promocion"].mean()
medio_mayor_impacto = impacto_por_medio.idxmax()
print("\nMedio publicitario con mayor impacto:", medio_mayor_impacto)

# d) Ubicación de residencia que más aficionados aporta
aficionados_por_ubicacion = df_completo.groupby("ubicación_residencia")["id_aficionado"].count()
ubicacion_mayor_afluencia = aficionados_por_ubicacion.idxmax()
print("\nUbicación de residencia que más aficionados aporta:", ubicacion_mayor_afluencia)

# Paso 6: Exportar DataFrame final a CSV
df_completo.to_csv("/workspace/reporte_eventos.csv", index=False)
print("\nArchivo 'reporte_eventos.csv' exportado correctamente.")
```
